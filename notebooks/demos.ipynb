# Random walks

```python
n, m = 1000, 500
paths = pd.DataFrame(np.random.randn(n, m)).cumsum()
paths.plot(legend=False, style='C0-', alpha=0.025, 
           title=f'Random walks\n{m} paths for {n} steps',
           xlim=0)
```
